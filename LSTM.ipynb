```
# ============================================================
# 1) Install / import
# ============================================================
!pip -q install tensorflow matplotlib scikit-learn pandas numpy

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix
import tensorflow as tf
from tensorflow.keras import layers, models

np.random.seed(42)
tf.random.set_seed(42)

print("TensorFlow version:", tf.__version__)

# ============================================================
# 2) Generate synthetic system activity data
#    No external dataset required
# ============================================================
# Features:
# 0 - cpu_usage
# 1 - memory_usage
# 2 - disk_io
# 3 - network_traffic
# 4 - process_count
# 5 - failed_logins

SEQ_LEN = 30
N_FEATURES = 6

def generate_normal_sequence(seq_len=30):
    t = np.arange(seq_len)

    cpu = 40 + 10*np.sin(2*np.pi*t/seq_len) + np.random.normal(0, 1.5, seq_len)
    memory = 55 + 6*np.sin(2*np.pi*t/seq_len + 0.8) + np.random.normal(0, 1.2, seq_len)
    disk_io = 30 + 5*np.sin(4*np.pi*t/seq_len) + np.random.normal(0, 1.5, seq_len)
    network = 45 + 8*np.sin(2*np.pi*t/seq_len + 1.2) + np.random.normal(0, 1.8, seq_len)
    process_count = 60 + 4*np.sin(2*np.pi*t/seq_len + 0.3) + np.random.normal(0, 1.0, seq_len)
    failed_logins = np.clip(np.random.poisson(0.3, seq_len), 0, None)

    seq = np.stack([cpu, memory, disk_io, network, process_count, failed_logins], axis=1)
    return seq

def inject_anomaly(seq):
    seq = seq.copy()
    anomaly_type = np.random.choice(["spike", "drift", "burst", "login_attack"])

    if anomaly_type == "spike":
        idx = np.random.randint(5, len(seq)-5)
        seq[idx:idx+3, 0] += np.random.uniform(25, 40)   # cpu spike
        seq[idx:idx+3, 3] += np.random.uniform(20, 35)   # network spike

    elif anomaly_type == "drift":
        start = np.random.randint(10, len(seq)-10)
        drift = np.linspace(0, np.random.uniform(15, 30), len(seq)-start)
        seq[start:, 1] += drift                           # memory drift
        seq[start:, 4] += drift * 0.3                     # process drift

    elif anomaly_type == "burst":
        idx = np.random.randint(3, len(seq)-6)
        seq[idx:idx+5, 2] += np.random.uniform(20, 35)    # disk burst
        seq[idx:idx+5, 3] += np.random.uniform(15, 30)    # traffic burst

    elif anomaly_type == "login_attack":
        idx = np.random.randint(5, len(seq)-5)
        seq[idx:idx+4, 5] += np.random.randint(8, 20)     # failed login spike
        seq[idx:idx+4, 3] += np.random.uniform(10, 20)    # network rise

    return seq

def generate_dataset(n_normal=1200, n_anomaly=300, seq_len=30):
    normal = np.array([generate_normal_sequence(seq_len) for _ in range(n_normal)])
    anomaly = np.array([inject_anomaly(generate_normal_sequence(seq_len)) for _ in range(n_anomaly)])

    X = np.concatenate([normal, anomaly], axis=0)
    y = np.concatenate([np.zeros(n_normal), np.ones(n_anomaly)], axis=0)  # 0=normal, 1=anomaly
    return X, y, normal, anomaly

X_all, y_all, X_normal, X_anomaly = generate_dataset(
    n_normal=1200,
    n_anomaly=300,
    seq_len=SEQ_LEN
)

print("All data shape:", X_all.shape)
print("Labels shape:", y_all.shape)
print("Normal sequences:", (y_all == 0).sum())
print("Anomalous sequences:", (y_all == 1).sum())

# ============================================================
# 3) Train / validation / test split
#    Train only on NORMAL sequences
# ============================================================
train_size = 800
val_size = 200

X_train = X_normal[:train_size]
X_val_normal = X_normal[train_size:train_size+val_size]

X_test = np.concatenate([
    X_normal[train_size+val_size:],
    X_anomaly
], axis=0)

y_test = np.concatenate([
    np.zeros(len(X_normal[train_size+val_size:])),
    np.ones(len(X_anomaly))
], axis=0)

print("X_train:", X_train.shape)
print("X_val_normal:", X_val_normal.shape)
print("X_test:", X_test.shape)
print("y_test:", y_test.shape)

# ============================================================
# 4) Normalize features
#    Fit scaler only on training-normal data
# ============================================================
scaler = StandardScaler()

X_train_2d = X_train.reshape(-1, N_FEATURES)
scaler.fit(X_train_2d)

def scale_3d(X, scaler):
    X_2d = X.reshape(-1, X.shape[-1])
    X_scaled = scaler.transform(X_2d)
    return X_scaled.reshape(X.shape)

X_train_scaled = scale_3d(X_train, scaler)
X_val_normal_scaled = scale_3d(X_val_normal, scaler)
X_test_scaled = scale_3d(X_test, scaler)

# ============================================================
# 5) Build LSTM Autoencoder
# ============================================================
inputs = layers.Input(shape=(SEQ_LEN, N_FEATURES))

# Encoder
x = layers.LSTM(64, activation="tanh", return_sequences=True)(inputs)
x = layers.LSTM(32, activation="tanh", return_sequences=False)(x)
latent = layers.Dense(16, activation="relu")(x)

# Decoder
x = layers.RepeatVector(SEQ_LEN)(latent)
x = layers.LSTM(32, activation="tanh", return_sequences=True)(x)
x = layers.LSTM(64, activation="tanh", return_sequences=True)(x)
outputs = layers.TimeDistributed(layers.Dense(N_FEATURES))(x)

model = models.Model(inputs, outputs)
model.compile(optimizer="adam", loss="mse")

model.summary()

# ============================================================
# 6) Train autoencoder on NORMAL sequences only
# ============================================================
early_stopping = tf.keras.callbacks.EarlyStopping(
    monitor="val_loss",
    patience=8,
    restore_best_weights=True
)

history = model.fit(
    X_train_scaled, X_train_scaled,
    validation_data=(X_val_normal_scaled, X_val_normal_scaled),
    epochs=50,
    batch_size=32,
    callbacks=[early_stopping],
    verbose=1
)

# ============================================================
# 7) Plot training history
# ============================================================
plt.figure(figsize=(8, 4))
plt.plot(history.history["loss"], label="Train Loss")
plt.plot(history.history["val_loss"], label="Validation Loss")
plt.title("LSTM Autoencoder Training Loss")
plt.xlabel("Epoch")
plt.ylabel("MSE")
plt.legend()
plt.grid(True)
plt.show()

# ============================================================
# 8) Reconstruction error
# ============================================================
def reconstruction_error(model, X):
    X_pred = model.predict(X, verbose=0)
    # Mean squared error per sequence
    error = np.mean(np.square(X - X_pred), axis=(1, 2))
    return error, X_pred

train_errors, X_train_pred = reconstruction_error(model, X_train_scaled)
val_errors, X_val_pred = reconstruction_error(model, X_val_normal_scaled)
test_errors, X_test_pred = reconstruction_error(model, X_test_scaled)

print("Train error stats:")
print(" mean =", train_errors.mean(), " std =", train_errors.std())
print("Val error stats:")
print(" mean =", val_errors.mean(), " std =", val_errors.std())
print("Test error stats:")
print(" mean =", test_errors.mean(), " std =", test_errors.std())

# ============================================================
# 9) Choose anomaly threshold
#    Common approach: threshold from normal validation error
# ============================================================
threshold = val_errors.mean() + 3 * val_errors.std()
print("Anomaly threshold:", threshold)

y_pred = (test_errors > threshold).astype(int)

# ============================================================
# 10) Evaluate anomaly detection
# ============================================================
print("\nClassification Report:")
print(classification_report(y_test, y_pred, digits=4))

print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))

# ============================================================
# 11) Visualize reconstruction error distribution
# ============================================================
normal_test_errors = test_errors[y_test == 0]
anomaly_test_errors = test_errors[y_test == 1]

plt.figure(figsize=(10, 5))
plt.hist(normal_test_errors, bins=40, alpha=0.7, label="Normal")
plt.hist(anomaly_test_errors, bins=40, alpha=0.7, label="Anomaly")
plt.axvline(threshold, color="red", linestyle="--", label="Threshold")
plt.title("Reconstruction Error Distribution")
plt.xlabel("Reconstruction Error")
plt.ylabel("Count")
plt.legend()
plt.grid(True)
plt.show()

# ============================================================
# 12) Plot some example reconstructions
# ============================================================
feature_names = [
    "cpu_usage", "memory_usage", "disk_io",
    "network_traffic", "process_count", "failed_logins"
]

def plot_reconstruction_example(X_true, X_recon, idx, title):
    fig, axes = plt.subplots(N_FEATURES, 1, figsize=(10, 12), sharex=True)
    for i in range(N_FEATURES):
        axes[i].plot(X_true[idx, :, i], label="Original")
        axes[i].plot(X_recon[idx, :, i], label="Reconstructed")
        axes[i].set_ylabel(feature_names[i])
        axes[i].grid(True)
        if i == 0:
            axes[i].set_title(title)
    axes[-1].set_xlabel("Time step")
    axes[0].legend()
    plt.tight_layout()
    plt.show()

# one normal example
normal_idx = np.where(y_test == 0)[0][0]
plot_reconstruction_example(X_test_scaled, X_test_pred, normal_idx, "Normal Sequence Reconstruction")

# one anomalous example
anomaly_idx = np.where(y_test == 1)[0][0]
plot_reconstruction_example(X_test_scaled, X_test_pred, anomaly_idx, "Anomalous Sequence Reconstruction")

# ============================================================
# 13) Show top anomalous sequences
# ============================================================
results = pd.DataFrame({
    "true_label": y_test.astype(int),
    "reconstruction_error": test_errors,
    "predicted_label": y_pred.astype(int)
})

results = results.sort_values("reconstruction_error", ascending=False).reset_index(drop=True)
print(results.head(10))

# ============================================================
# 14) Optional: inspect a few samples
# ============================================================
plt.figure(figsize=(10, 4))
plt.plot(test_errors[:100], marker='o', linestyle='-', alpha=0.8)
plt.axhline(threshold, color="red", linestyle="--", label="Threshold")
plt.title("Reconstruction Error for First 100 Test Sequences")
plt.xlabel("Sequence Index")
plt.ylabel("Error")
plt.legend()
plt.grid(True)
plt.show()
```
